```python
import pandas as pd
import numpy as np

volume_csv = "/NFS/Wendy/homework/HW2/output/volume_calculations.csv"
demographics_path = "/NFS/Wendy/homework/HW2/IXI.csv"

volume_df = pd.read_csv(volume_csv)
demographics_df = pd.read_csv(demographics_path)

if 'DOB' in demographics_df.columns:
    demographics_df['DOB'] = pd.to_datetime(demographics_df['DOB'], errors='coerce')
    demographics_df['AGE'] = (2024 - demographics_df['DOB']).dt.year

demographics_df.drop(columns=['DOB'], inplace=True)

merged_df = pd.merge(volume_df, demographics_df, on="IXI_ID")

labels = {
    1: "Left-Thalamus-Proper", 
    2: "Right-Thalamus-Proper",
    3: "Left-Caudate", 
    4: "Right-Caudate",
    5: "Left-Putamen", 
    6: "Right-Putamen",
    7: "Left-Pallidum", 
    8: "Right-Pallidum",
    9: "Left-Hippocampus", 
    10: "Right-Hippocampus",
    11: "Left-Amygdala", 
    12: "Right-Amygdala"
}

variables = ['AGE', 'SEX_ID', 'HEIGHT', 'WEIGHT', 'ETHNIC_ID']

correlation_data = []

for key, structure in labels.items():
        structure_data = {'Structure': structure}
        
        for var in variables:
            filtered_data = merged_df[[var, structure]].dropna()
            
            if not filtered_data.empty:
                correlation = filtered_data[var].corr(filtered_data[structure])
                structure_data[f'{var}-Volume Correlation'] = correlation if not np.isnan(correlation) else 'N/A'
            else:
                structure_data[f'{var}-Volume Correlation'] = 'N/A'
        
        correlation_data.append(structure_data)

correlation_report_df = pd.DataFrame(correlation_data)
print("Correlation Analysis Report Across All Patients:")
print(correlation_report_df)

output = '/NFS/Wendy/homework/HW2/volume_correlation_report.csv'
correlation_report_df.to_csv(output, index=False)
print(f"Correlation report saved to {output}")
```

```python
import nibabel as nib
import numpy as np
import pandas as pd
import glob
import os

gm_file = "/NFS/Wendy/homework/HW2/graymatter/*.nii.gz"
output =  "/NFS/Wendy/homework/HW2/volume_results.csv"
gm_files = glob.glob(gm_file)

volume_results = {}

labels = {
    1: "Left-Thalamus-Proper", 
    2: "Right-Thalamus-Proper",
    3: "Left-Caudate", 
    4: "Right-Caudate",
    5: "Left-Putamen", 
    6: "Right-Putamen",
    7: "Left-Pallidum", 
    8: "Right-Pallidum",
    9: "Left-Hippocampus", 
    10: "Right-Hippocampus",
    11: "Left-Amygdala", 
    12: "Right-Amygdala"
}

for gm_file_all in gm_files:
    segmentation_img = nib.load(gm_file_all)
    segmentation_data = segmentation_img.get_fdata()
    
    s = segmentation_img.header.get_zooms()
    voxel_volume = np.prod(s)
    
    volumes = {}
    for label, region_name in labels.items():
        region_voxels = np.sum(segmentation_data == label)
        volumes[region_name] =round(region_voxels * voxel_volume, 1)
        
        short = os.path.basename(gm_file_all)[:6]  # 取前六個字母
        volume_results[short] = volumes
        

volume_df = pd.DataFrame(volume_results).T
volume_df.to_csv(output, index_label= "IXI_ID")

print(volume_df)
```

```python
import pandas as pd
import numpy as np

volume_result = "/NFS/Wendy/homework/HW2/volume_results.csv"
IXI_path = "/NFS/Wendy/homework/HW2/IXI.csv"

volume_file = pd.read_csv(volume_result)
IXI_file = pd.read_csv(IXI_path)


volume_file['IXI_ID'] = volume_file['IXI_ID'].str.extract(r'(\d+)').astype(int)
merged_file = pd.merge(volume_file, IXI_file, on="IXI_ID")
merged_file = merged_file[(merged_file['HEIGHT'] <= 500) & (merged_file['WEIGHT'] <= 500)]

labels = {
    1: "Left-Thalamus-Proper", 
    2: "Right-Thalamus-Proper",
    3: "Left-Caudate", 
    4: "Right-Caudate",
    5: "Left-Putamen", 
    6: "Right-Putamen",
    7: "Left-Pallidum", 
    8: "Right-Pallidum",
    9: "Left-Hippocampus", 
    10: "Right-Hippocampus",
    11: "Left-Amygdala", 
    12: "Right-Amygdala"
}

variables = ['SEX_ID', 'HEIGHT', 'WEIGHT', 'ETHNIC_ID', 'AGE']

correlation_data = []

for label, region_name in labels.items():
    region_data = {' region': region_name}
    
    for var in variables:
        filtered_data = merged_file[[var, region_name]].dropna()
        
        if not filtered_data.empty:
            correlation = filtered_data[var].corr(filtered_data[region_name])
            region_data[f'{var}-Volume Correlation'] = round(correlation, 3) if not np.isnan(correlation) else 'N/A'
            
        else:
            region_data[f'{var}-Volume Correlation'] = 'N/A'
    correlation_data.append(region_data)

correlation_report_df = pd.DataFrame(correlation_data)
print("Correlation Analysis Report Across All Patients:")
print(correlation_report_df)

output = '/NFS/Wendy/homework/HW2/volume_correlation_report4.csv'
correlation_report_df.to_csv(output, index=False)
```

```python
import pandas as pd

file1 = pd.read_csv('IXI.csv')
file2 = pd.read_csv('volume_results.csv')

file1['IXI_ID'] = file1['IXI_ID'].apply(lambda x: f"IXI{x:03}")

merged_data = pd.merge(file2, file1[['IXI_ID', 'AGE']], on='IXI_ID', how='left')
merged_data = merged_data.dropna(subset=['AGE'])

merged_data.to_csv('merge_data.csv', index=False)
print(merged_data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('merge_data.csv')

if data.isnull().values.any():
    data = data.dropna()


x = data[['Left-Thalamus-Proper','Right-Thalamus-Proper','Left-Caudate',
          'Right-Caudate','Left-Putamen', 'Right-Putamen','Left-Pallidum', 
          'Right-Pallidum', 'Left-Hippocampus', 'Right-Hippocampus',
          'Left-Amygdala', 'Right-Amygdala']]
y = data['AGE']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"MAE:{mae}")
print(f"MSE:{mse}")

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.xlabel('Actual Age')
plt.ylabel('Predict Age')
plt.show
```
